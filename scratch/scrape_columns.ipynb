```python
import json
import requests
import urllib.request
from bs4 import BeautifulSoup


def scrape_columns():
    
    base_path = 'https://api.census.gov/data/2023/cps/asec/mar/'
    listing = 'https://api.census.gov/data/2023/cps/asec/mar/variables.html'
    content = requests.get(listing)
    soup = BeautifulSoup(content.text)
    
    # find the table body and then find all anchor links
    tbody = soup.find_all("tbody".split())[0]
    anchortags = tbody.find_all('a', href=True)[::2]
    
    # prepare a dictionary for each attribute
    result = {}
    print(f"len anchor tags = {len(anchortags)}")
    
    for i, anchortag in enumerate(anchortags):
        link = anchortag['href']
        if link == "/data/2023/cps/asec/mar/variables/.html":
            continue
        url1 = base_path + link
#         print(url1)
        
        if (i+1)%50==0:
            print(f"Fetching {i+1}/{len(anchortags)}")
        with urllib.request.urlopen(url1) as url:
            data = json.load(url)
            result[data['name']] = data
            
#         if i==10:
#             break
            
    return result
```

```python
attributes = scrape_columns()
```
